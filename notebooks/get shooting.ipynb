```python
elem = ps.driver.find_element_by_id('shooting').text

df = pd.read_csv(StringIO(elem), sep=" ").reset_index()

df.columns = df.iloc[0]
df = df.drop(0)

df.columns

lvl0 = [col for col in df.columns[:9]]

lvl1 = df.columns[11:17] #% of FGA by Distance

lvl1

lvl1 = [f'{col}_%FGAbyDistance' for col in df.columns[11:17]]
lvl1

lvl2 = df.columns[19:25]
lvl2 = [f'{col}_FG%byDistance' for col in df.columns[19:25]]#FG% by distance
lvl2

lvl3 = df.columns[27:29]
lvl3 = [f'{col}_%FGAst' for col in lvl3] #% of FG Ast'd
lvl3

lvl4 = df.columns[31:33]
lvl4 = [f'{col}_Dunks' for col in lvl4] 
lvl4#Dunks

lvl5 = df.columns[35:37] #Corner Threes
lvl5 = [f'{col}_CornerThrees' for col in lvl5]
lvl5

lvl6 = df.columns[39:41] #Heaves
lvl6 = [f'{col}_Heaves' for col in lvl6]
lvl6

shooting_columns = lvl0+lvl1+lvl2+lvl3+lvl4+lvl5+lvl6

shooting_df = df.dropna(how = 'all', axis = 1)
shooting_df.columns = shooting_columns

shooting_df = shooting_df.set_index('Season')

shooting_df.iloc[-3:] = shooting_df.iloc[-3:].shift(1, axis = 1)

shooting_df.iloc[-4] = shooting_df.iloc[-4].shift(3)

shooting_df.reset_index()
```
